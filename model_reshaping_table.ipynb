```python
# prompt: I want to read each sheets in above uploaded excel file Dt_features_coef.xlsx go through each sheet and drop the columns with names  std err , t , P>|t| , [0.025 . 0.975]

import pandas as pd
excel_file = '/content/drive/MyDrive/Flipose-Analytics/Work/Vismaya/Dt_features_coef.xlsx'
xls = pd.ExcelFile(excel_file)

# List to store dataframes after dropping columns
processed_dfs = {}

# Columns to drop
columns_to_drop = ['std err', 't', 'P>|t|', '[0.025', '0.975]']

for sheet_name in xls.sheet_names:
    # Read each sheet into a pandas DataFrame
    df = pd.read_excel(xls, sheet_name)

    # Drop specified columns if they exist
    cols_to_drop_exist = [col for col in columns_to_drop if col in df.columns]
    if cols_to_drop_exist:
        df_processed = df.drop(columns=cols_to_drop_exist)
    else:
        df_processed = df.copy() # Keep original if no columns to drop

    # Store the processed DataFrame
    processed_dfs[sheet_name] = df_processed

# Now, processed_dfs is a dictionary where keys are sheet names and values are
# the DataFrames with the specified columns dropped.
# You can now work with these processed DataFrames.
# For example, to see the first few rows of the processed 'Sheet1':
# print(processed_dfs['Sheet1'].head())

# Or you can save the processed dataframes to a new Excel file
output_file = '/content/drive/MyDrive/Flipose-Analytics/Work/Vismaya/Dt_features_coef_processed.xlsx'
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    for sheet_name, df_processed in processed_dfs.items():
        df_processed.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"Processed sheets saved to: {output_file}")
```

```python
# prompt: Now I want to transpose variable column to rows with their values add to above code

import pandas as pd
excel_file = '/content/drive/MyDrive/Flipose-Analytics/Work/Vismaya/Dt_features_coef_processed.xlsx'
xls = pd.ExcelFile(excel_file)

# List to store transposed dataframes
transposed_dfs = {}

for sheet_name in xls.sheet_names:
    # Read each sheet into a pandas DataFrame
    df = pd.read_excel(xls, sheet_name)

    # Set 'Variable' as index for transposition
    df_indexed = df.set_index('Variable')

    # Transpose the dataframe
    df_transposed = df_indexed.T

    # Reset index and rename the new index column (optional)
    df_transposed = df_transposed.reset_index()
    df_transposed = df_transposed.rename(columns={'index': 'OriginalColumn'})

    # Store the transposed DataFrame
    transposed_dfs[sheet_name] = df_transposed

# Now, transposed_dfs is a dictionary where keys are sheet names and values are
# the transposed DataFrames.

# You can save the transposed dataframes to a new Excel file
output_file_transposed = '/content/drive/MyDrive/Flipose-Analytics/Work/Vismaya/Dt_features_coef_transposed.xlsx'
with pd.ExcelWriter(output_file_transposed, engine='xlsxwriter') as writer:
    for sheet_name, df_transposed in transposed_dfs.items():
        df_transposed.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"Transposed sheets saved to: {output_file_transposed}")
```
