```python
# prompt: add grou[_name on x-axis and nAreas and nRecords on y1-axis as bar chart and cum%_records and Cum%_areas on y2-axis and give different colours for each and keep legends on top of chart

import matplotlib.pyplot as plt
# Ensure the data is sorted by group_name for cumulative calculations if needed,
# though the provided file seems pre-calculated.
# Assuming 'group_name' is already sorted as per the image.

# Set a larger figure size for better readability
plt.figure(figsize=(14, 7))

# Create the primary y-axis (left) for bars
ax1 = sns.barplot(x='Group_name', y='%Area', data=df, color='skyblue', label='%Area')
sns.barplot(x='Group_name', y='%Records', data=df, color='lightcoral', label='%Records', ax=ax1)

# Create the secondary y-axis (right) for lines
ax2 = ax1.twinx()
sns.lineplot(x='Group_name', y='Cum%_records', data=df, color='green', marker='o', label='Cum%_records', ax=ax2)
sns.lineplot(x='Group_name', y='Cum%_areas', data=df, color='darkorange', marker='o', label='Cum%_areas', ax=ax2)

# Set axis labels
ax1.set_xlabel('Group_name')
ax1.set_ylabel('Counts (nAreas, nRecords)')
ax2.set_ylabel('Cumulative Percentage')

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=4)

# Set the title of the plot
plt.title('ABC Analysis Summary')

# Improve layout to prevent labels overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add bar charts for %Area and %Records
fig.add_trace(
    go.Bar(name='%Area', x=df['Group_name'], y=df['%Area'], marker_color='skyblue',
           hovertemplate='<b>%{x}</b><br>%Area: %{y:.2f}%<extra></extra>'),
    secondary_y=False,
)

fig.add_trace(
    go.Bar(name='%Records', x=df['Group_name'], y=df['%Records '], marker_color='lightcoral',
           hovertemplate='<b>%{x}</b><br>%Records: %{y:.2f}%<extra></extra>'),
    secondary_y=False,
)

# Add line charts for cumulative percentages
fig.add_trace(
    go.Scatter(name='Cum%_records', x=df['Group_name'], y=df['Cum%_records'], mode='lines+markers', marker_color='green',
               hovertemplate='<b>%{x}</b><br>Cum% Records: %{y:.2f}%<extra></extra>'),
    secondary_y=True,
)

fig.add_trace(
    go.Scatter(name='Cum%_areas', x=df['Group_name'], y=df['Cum%_areas'], mode='lines+markers', marker_color='darkorange',
               hovertemplate='<b>%{x}</b><br>Cum% Areas: %{y:.2f}%<extra></extra>'),
    secondary_y=True,
)

# Set x-axis title
fig.update_xaxes(title_text='Group_name')

# Set y-axes titles
fig.update_yaxes(title_text='Counts (%Area, %Records)', secondary_y=False)
fig.update_yaxes(title_text='Cumulative Percentage', secondary_y=True)

# Add title and legend
fig.update_layout(
    title_text='ABC Analysis Summary',
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    hovermode='x unified' # Set hover mode to show all trace values at a given x-coordinate
)

# Show the plot
fig.show()
```

```python
# prompt: I want area_name_en in x-axis and nRecords on y1-axis and Cumulative_% on y2-axis and add lines at  Wadi Al Safa 5 as 40% and Al Hebiah Third as70%(breakdown line) on cumlative% paralell to y axis and I want y1-axis scale as 0 ,20k,40k,60k ...100k and I want to increase the height of chart

import numpy as np
# Ensure the dataframe is sorted by nRecords in descending order for cumulative calculations
df2_sorted = df2.sort_values(by='nRecords', ascending=False).reset_index(drop=True)

# Calculate cumulative sum of nRecords and cumulative percentage
df2_sorted['Cumulative_nRecords'] = df2_sorted['nRecords'].cumsum()
df2_sorted['Cumulative_%'] = (df2_sorted['Cumulative_nRecords'] / df2_sorted['nRecords'].sum()) * 100

# Create figure with secondary y-axis and increased height
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add bar chart for nRecords on the primary y-axis
fig.add_trace(
    go.Bar(name='nRecords', x=df2_sorted['area_name_en'], y=df2_sorted['nRecords'], marker_color='blue',
           hovertemplate='<b>%{x}</b><br>nRecords: %{y}<extra></extra>'),
    secondary_y=False,
)

# Add line chart for Cumulative_% on the secondary y-axis
fig.add_trace(
    go.Scatter(name='Cumulative_%', x=df2_sorted['area_name_en'], y=df2_sorted['Cumulative_%'], mode='lines', marker_color='red',
               hovertemplate='<b>%{x}</b><br>Cumulative %: %{y:.2f}%<extra></extra>'),
    secondary_y=True,
)

# Set x-axis title
fig.update_xaxes(title_text='area_name_en')

# Set y-axes titles
fig.update_yaxes(title_text='nRecords', secondary_y=False)
fig.update_yaxes(title_text='Cumulative %', secondary_y=True)

# Set the y1-axis ticks
y1_max = df2_sorted['nRecords'].max()
y1_ticks = np.arange(0, y1_max * 1.1, 20000) # Adjust max and step as needed
fig.update_yaxes(tickvals=y1_ticks, secondary_y=False)


# Find the index for 'Wadi Al Safa 5' and 'Al Hebiah Third'
wadi_safa_index = df2_sorted[df2_sorted['area_name_en'] == 'Wadi Al Safa 5'].index
al_hebiah_index = df2_sorted[df2_sorted['area_name_en'] == 'Al Hebiah Third'].index

# Add vertical lines at the specified area names based on their index
if not wadi_safa_index.empty:
    fig.add_vline(x=wadi_safa_index[0], line_dash="dash", line_color="green")
if not al_hebiah_index.empty:
    fig.add_vline(x=al_hebiah_index[0], line_dash="dash", line_color="purple")




# Add title and legend, and update layout height
fig.update_layout(
    title_text='Pareto Analysis by Area',
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    hovermode='x unified', # Set hover mode to show all trace values at a given x-coordinate
    height=800, # Increase the height of the chart
    barmode='group' # Ensure bars are grouped if needed, though only one bar trace here
)

# Show the plot
fig.show()
```
